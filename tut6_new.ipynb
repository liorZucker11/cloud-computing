```python
from firebase import firebase
firebase = firebase.FirebaseApplication('https://cloudtut6-default-rtdb.europe-west1.firebasedatabase.app/', None)
index = firebase.get('/frequent_words',None)
print(index)
```

```python
def search(word):
  word_data = firebase.get(f'/frequent_words/{word}',None)
  return word_data['links']
```

```python
def count(word):
  word_data = firebase.get(f'/frequent_words/{word}',None)
  return word_data['count']
```

```python
print('Azure',search('Azure'))
print('cloud',search('cloud'))
print('storage',search('storage'))
print('management',search('management'))
print('pricing',search('pricing'))
print('support',search('support'))
print('security',search('security'))
print('development',search('development'))
print('SQL',search('SQL'))
print('migration',search('migration'))
```

```python
import matplotlib.pyplot as plt

words = []
for item in index:
  words.append(item)

counts = []
for item in index:
  counts.append(index[item]['count'])

plt.figure(figsize=(10, 6))
plt.bar(words, counts, color='skyblue')
plt.xlabel('Words')
plt.ylabel('Counts')
plt.xticks(rotation=45, ha="right")
plt.title('Word Counts in Links')
plt.tight_layout()
plt.show()
```
