```python
from firebase import firebase
firebase = firebase.FirebaseApplication('https://cloudtut6-default-rtdb.europe-west1.firebasedatabase.app/', None)
index = firebase.get('/frequent_words',None)
```

```python
def search(word,index):
  links = None
  for item in index:
      if word in item:
          links = item[word]['links']
          break
  return links
```

```python
def count(word,index):
  count = None
  for item in index:
      if word in item:
          links = item[word]['count']
          break
  return count
```

```python
def print_links(word, links):
  print(f"{word}, links: {links}")
```

```python
print_links('Azure',search('Azure',index))
print_links('cloud',search('cloud',index))
print_links('storage',search('storage',index))
print_links('management',search('management',index))
print_links('pricing',search('pricing',index))
print_links('support',search('support',index))
print_links('security',search('security',index))
print_links('development',search('development',index))
print_links('SQL',search('SQL',index))
print_links('migration',search('migration',index))
```

```python
import matplotlib.pyplot as plt

words = [list(item.keys())[0] for item in index]
counts = [list(item.values())[0]["count"] for item in index]

plt.figure(figsize=(10, 6))
plt.bar(words, counts, color='skyblue')
plt.xlabel('Words')
plt.ylabel('Counts')
plt.xticks(rotation=45, ha="right")
plt.title('Word Counts in Links')
plt.tight_layout()
plt.show()
```
